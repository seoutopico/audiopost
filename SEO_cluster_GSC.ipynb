```python
# Instalar librer√≠as necesarias
!pip install hdbscan scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer
import hdbscan
import pandas as pd
from google.colab import files
```

```python
# Cargar el archivo CSV desde tu ordenador Queries.csv previamente descargado de Google search console
uploaded = files.upload()

# Leer el archivo CSV (cambia el nombre del archivo si es necesario)
df = pd.read_csv(list(uploaded.keys())[0])
```

```python
# Convertir las queries a TF-IDF (optimizando para queries cortas)
vectorizer = TfidfVectorizer(ngram_range=(1, 2))  # ngram_range para tener en cuenta pares de palabras
tfidf_matrix = vectorizer.fit_transform(df['Top queries'])

# Aplicar HDBSCAN con un tama√±o m√≠nimo de cluster ajustado a 5
clusterer = hdbscan.HDBSCAN(min_cluster_size=5, metric='euclidean', cluster_selection_method='eom')
cluster_labels = clusterer.fit_predict(tfidf_matrix.toarray())

# Agregar los clusters al DataFrame
df['Cluster'] = cluster_labels

# Guardar el resultado en un archivo CSV
output_file = 'clustered_queries_hdbscan_tfidf.csv'
df.to_csv(output_file, index=False)

# Descargar el archivo CSV
files.download(output_file)

# Imprimir el n√∫mero total de clusters creados y mostrar los primeros 5 clusters
total_clusters = df['Cluster'].nunique()
print(f"Total clusters creados: {total_clusters}")

# Mostrar los primeros 5 clusters
clusters_summary = {}
for i in range(5):
    clusters_summary[f"Cluster {i + 1}"] = df[df['Cluster'] == i]['Top queries'].tolist()

# Mostrar resumen de los primeros 5 clusters
print(clusters_summary)

# Funci√≥n para mostrar un resumen de los primeros 5 clusters
def mostrar_resumen_clusters(df, num_clusters=5):
    # Agrupar las queries por cluster
    clusters_summary = df.groupby('Cluster')['Top queries'].apply(list)

    print(f"Total clusters creados: {len(clusters_summary)}\n")

    # Mostrar solo los primeros N clusters (por defecto, 5)
    for i, (cluster, queries) in enumerate(clusters_summary.items()):
        if i >= num_clusters:
            break
        print(f"Cluster {cluster}:")
        for query in queries:
            print(f"  - {query}")
        print("\n")

# Mostrar el resumen de los primeros 5 clusters
mostrar_resumen_clusters(df, num_clusters=5)

# Guardar el resultado en un archivo CSV
df.to_csv('clustered_queries.csv', index=False)
```
